Dados a serem utilizados:

```
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```
loja1.head()
```

Cálculo do faturamento das lojas

```
loja1['Preço'].sum()
```

```
faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4)
```

Média

```
avaliacao_loja1 = loja1['Avaliação da compra'].mean()
avaliacao_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(avaliacao_loja1, avaliacao_loja2, avaliacao_loja3, avaliacao_loja4)
```

Frete

```
frete_loja1 = loja1['Frete'].mean()
frete_loja2 = loja2['Frete'].mean()
frete_loja3 = loja3['Frete'].mean()
frete_loja4 = loja4['Frete'].mean()

print(frete_loja1, frete_loja2, frete_loja3, frete_loja4)
```

Categoria do produto mais vendido

```
loja1['Categoria do Produto'].value_counts()
```

Gráficos para visualização

```
import matplotlib.pyplot as plt

lojas = ['Loja 1','Loja 2','Loja 3','Loja 4']
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.bar(lojas, faturamento)
plt.title('Faturamento por loja')
plt.show()
```

```
import matplotlib.pyplot as plt

categorias = loja1['Categoria do Produto'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%')
plt.title('Distribuição de categorias - Loja 1')
plt.show()
```

```
plt.figure(figsize=(6,4))
plt.scatter(loja1['Preço'], loja1['Avaliação da compra'])
plt.xlabel('Preço')
plt.ylabel('Avaliação')
plt.title('Relação entre preço e avaliação - Loja 1')
plt.show()
```
